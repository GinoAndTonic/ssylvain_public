```python
#Here we extract the tips and treas yield data and save the different html pages in a dictionary
i = 0
tips_dict ={}
while True:
    i += 1
    url ='https://www.federalreserve.gov/econresdata/researchdata/feds200805_'+str(i)+'.html'
    try:
        tips_dict['page'+str(i)] = pd.read_html(url)[-1]
    except:
        break
    print 'finished loading the following url:'    
    print url    
    print '/n'   
print 'finished loading all urls'    
```

```python
tips_dict['page1'].head()
```

```python
i = 0
nominal_dict ={}
while True:
    i += 1
    url ='http://www.federalreserve.gov/econresdata/researchdata/feds200628_'+str(i)+'.html'
    try:
        nominal_dict['page'+str(i)] = pd.read_html(url)[-1]
    except:
        break
    print 'finished loading the following url:'
    print url     
    print '/n'   
print 'finished loading all urls'
```

```python
nominal_dict['page1'].head()
```

```python
tips_data = pd.DataFrame(np.array(tips_dict['page1'].values[1:,1:],dtype=np.float64),
                         index=tips_dict['page1'].values[1:,0],columns= tips_dict['page1'].loc[0,1:].values)

for i in range(1,len(tips_dict.keys())):
    print 'appending page '+str(i)
    tips_data = tips_data.append(pd.DataFrame(np.array(tips_dict['page'+str(i)].values[1:,1:],dtype=np.float64),
                         index=tips_dict['page'+str(i)].values[1:,0],columns= tips_dict['page'+str(i)].loc[0,1:].values))
```
